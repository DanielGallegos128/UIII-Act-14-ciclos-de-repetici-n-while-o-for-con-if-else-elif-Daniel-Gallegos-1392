### Problema NO. 23 Invertir Numeros: Algoritmo que invierta un número entero introducido por teclado (ej: 1234 → 4321) sin convertirlo a texto.

```python
# Solicitar el número entero al usuario
numero = int(input("Introduce un número entero: "))

# Variables para el proceso
numero_original = numero  # Guardamos el original por si se necesita
numero_invertido = 0

# Manejo de números negativos para que la matemática funcione igual
es_negativo = False
if numero < 0:
    es_negativo = True
    numero = -numero

# Bucle para invertir el número matemáticamente
while numero > 0:
    # 1. Extraer el último dígito con el residuo de la división entre 10
    digito = numero % 10

    # 2. Desplazar los dígitos ya invertidos a la izquierda y sumar el nuevo
    numero_invertido = (numero_invertido * 10) + digito

    # 3. Eliminar el último dígito del número original con división entera
    numero //= 10

# Si el número original era negativo, devolvemos el signo
if es_negativo:
    numero_invertido = -numero_invertido

# Mostrar el resultado
print("Número invertido:", numero_invertido)
```
